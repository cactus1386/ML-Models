```python
import torch
from torch import nn
from torchvision.datasets import CIFAR10
from torchvision import transforms
from torch.utils.data import DataLoader
from torch.optim import SGD
```

```python
transform = transforms.Compose(
    [transforms.ToTensor(),])

dataset = CIFAR10("./", train=True, transform=transform, download=True)
```

```python
loader = DataLoader(dataset, batch_size=128)
```

```python
class MyNetwork(nn.Module):
  def __init__(self):
    super().__init__()
    self.layer_1 = nn.Linear(3 * 32 * 32, 512)
    self.layer_2 = nn.Linear(512, 256)
    self.layer_3 = nn.Linear(256, 128)
    self.layer_4 = nn.Linear(128, 10)

  def forward(self, x):
    x_flat = x.view(x.shape[0], -1)

    y1 = self.layer_1(x_flat)
    y2 = self.layer_2(y1)
    y3 = self.layer_3(y2)
    y4 = self.layer_4(y3)

    return y4

network = MyNetwork()
```

```python
opt = SGD(network.parameters(), lr=0.001)
loss_func = nn.CrossEntropyLoss()

for batch in loader:
   inputs = batch[0]
   targets = batch[1]

   y = network(inputs)

   loss_value = loss_func(y, targets)

   loss_value.backward()

   opt.step()

   print(loss_value.item())
```
