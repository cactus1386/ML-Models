```python
from torch import nn

class MyNetwork(nn.Module):
  def __init__(self):
    super().__init__()
    self.layer_1 = nn.Linear(4,5)
    self.layer_2 = nn.Linear(5, 10)
    self.layer_3 = nn.Linear(10, 3)


  def forward(self, x):
    y1 = self.layer_1(x)
    y2 = self.layer_2(y1)
    y3 = self.layer_3(y2)

    return y3
```

```python
import torch
from torch.optim import SGD
from torch.nn import MSELoss

X = torch.rand(100, 4)
Y = torch.rand(100, 3)

network = MyNetwork()
opt = SGD(network.parameters(), lr=0.001)
loss_func = MSELoss()

for i in range(1000):
  y = network(X)
  loss_value = loss_func(y, Y)

  print(f"The Value of loss function is {loss_value.item()}")

  loss_value.backward()
  opt.step()
```
