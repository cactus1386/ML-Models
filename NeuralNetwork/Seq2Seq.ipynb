```python
from torch import nn
import torch
```

```python
inputs = torch.rand(10, 3, 512)  # Its make 3 sequence with 10 lenght and 512 feature

encoder_layer = nn.TransformerEncoderLayer(d_model=512, nhead=4)
encoder = nn.TransformerEncoder(encoder_layer, num_layers=5)

output = encoder(inputs)
output.size()
```

# Masked Language Model

```python
from random import random

masked_inputs = inputs

for i in range(3):
  for j in range(10):
    if random() < 0.15: # Just 15 percent of this loop
      masked_inputs[j, i].data = torch.zeros(512)

loss_func = nn.MSELoss()

outputs = encoder(masked_inputs)

loss_value = loss_func(outputs, inputs)
loss_value.item()
```

# Next Token Predict

```python
mask = torch.zeros(10, 10)

for i in range(10):
  for j in range(i + 1, 10):
    mask[j, i] = float('-inf')

outputs = encoder(inputs, src_mask=mask)

loss_func = nn.MSELoss()
loss_value = loss_func(outputs, inputs)
```

# Seq2Seq

```python
decoder_layer = nn.TransformerDecoderLayer(d_model=512, nhead=4)
decoder = nn.TransformerDecoder(decoder_layer, 10)

encoder_output = encoder(inputs)
target = torch.rand(8, 3, 512)

decoder_outputs = decoder(target, encoder_output, src_mask=mask)

loss_value = loss_func(decoder_outputs, target)
```
