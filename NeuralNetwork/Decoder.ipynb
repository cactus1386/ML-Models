```python
class Tokenizer():
  def __init__(self):
    self.words = [
        "<bos>",
        "<eos>",
        "<pad>",
        "<unk>",
        "سلام",
        "چطوری",
        "فداتشم",
        "بدی",
        "خوبی",
        "فعلا",
        "چخبرا"
    ]

    self.indexs = {
        word: id for id, word in enumerate(self.words)
    }

    self.eos = 1

  def __len__(self):
    return len(self.words)

  def id_to_word(self, indexs: list[int]) -> str:
    pass

  def word_to_id(self, sentence: str) -> list[int]:
    words = sentence.split(' ')
    results = []
    for word in words:
      results.append(self.indexs[word])

    return results
```

```python
import torch
from torch import nn

class Decoder(nn.Module):
  def __init__(self, tokenizer: Tokenizer):
    super().__init__()
    self.tokenizer = tokenizer
    self.embedding = nn.Embedding(len(tokenizer), 128)

    self.rnn = nn.GRU(128, 256, 4)

    self.linear = nn.Linear(256, len(tokenizer))

    def forward(self, x, h):
      emb = self.embedding(x)
      next_tokens = []
      while True:
        current_token = emb
        next_feature, h = self.rnn(current_token, h)
        next_token = self.linear(next_feature)
        next_token = torch.argmax(next_token, dim=-1)
        next_tokens.append(next_token)
        emb = self.embedding(next_token)

        if next_token == self.tokenizer.eos:
          break

      return next_tokens
```
