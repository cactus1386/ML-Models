```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

Generate data use "make blobs" function

```python
num_samples = 1000
cluster_centers = [(3,3), (7,7)]
num_classes = len(cluster_centers)
```

```python
X, y = make_blobs(n_samples=num_samples, centers=cluster_centers, n_features=num_classes, center_box=(0,1), cluster_std=0.5)
```

Use DBSCAN model

```python
db = DBSCAN(eps=0.5, min_samples=20).fit(X)
labels = db.labels_
```

Number of cluster and noiseData

```python
n_clusters = len(np.unique(labels))
n_noise = np.sum(np.array(labels) == -1, axis=0)

print("Number of clusters: %d" % n_clusters)
print("Number of noise points: %d" % n_noise)
```

Plot Data

```python
colors = list(map(lambda x: '#3b4cc0' if x == 0 else ("#b40426" if x == 1 else '#000000'),labels))
plt.scatter(X[:,0], X[:,1], c=colors, marker='o', picker=True)
plt.show()
```
