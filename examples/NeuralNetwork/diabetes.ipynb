```python
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score, r2_score
```

```python
data = load_diabetes()
x, y = data.data, data.target
x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
scaler = StandardScaler()
x_trainscaled = scaler.fit_transform(x_train)
x_testscaled = scaler.fit_transform(x_test)
```

```python
MLP = MLPRegressor(hidden_layer_sizes=(300, 400, 400, 300, 300, 200, 200, 200, 50, 50), activation='relu', max_iter=400,
    random_state=42, verbose=True, learning_rate='adaptive', learning_rate_init=0.001)
```

```python
MLP.fit(x_trainscaled, y_train)
```

```python
MLP.predict(x_testscaled)
MLP.score(x_testscaled, y_test)
```

```python
y_pred = MLP.predict(x_testscaled)
r2_score(y_pred=y_pred, y_true=y_test)
```
